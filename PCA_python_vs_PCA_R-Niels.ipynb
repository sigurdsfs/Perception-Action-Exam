```python
pca = PCA()
std_scal = StandardScaler()
df_pos_all_std = std_scal.fit_transform(df_pos_all)
df_pos_all_pca = pca.fit_transform(df_pos_all_std)
mean_pos_all = df_pos_all_pca.mean(axis = 0)
mean_df_pca_them = df_pca_them.mean(axis = 0)
plt.plot(mean_pos_all[0:14], color = 'b')
plt.plot(np.asarray(mean_df_pca_them)[0:14], color = 'r')
```

```python
pca = PCA()
df_pos_all_std = np.zeros(df_pos_all.shape)
for i in range(df_pos_all.shape[1]):
    std_scal = StandardScaler()
    df_pos_all_std[:,i] = np.squeeze(std_scal.fit_transform(df_pos_all.values[:,i].reshape(-1,1)))
```

```python
    
df_pos_all_pca = pca.fit_transform(df_pos_all_std)
mean_pos_all = df_pos_all_pca.mean(axis = 0)
mean_df_pca_them = df_pca_them.mean(axis = 0)
plt.plot(mean_pos_all[0:14], color = 'b')
plt.plot(np.asarray(mean_df_pca_them)[0:14], color = 'r')
```
